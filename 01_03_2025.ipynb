```python
import re
from datasets import Dataset
raw_text="""
Hugging Face!!! specializes in NLP... It's great platform for working with predefined models.
Visit: https://huggingface.co/ for more info.
Contact us at support@huggingface.co.
"""
def clean_text(text):
  text=text.lower()
  text=re.sub(r'https?://\S+','',text)
  text=re.sub(r'\S+@\S+','',text)
  text=re.sub(r'[^a-zA-z0-9\s]','',text)
  text=re.sub(r'\s+',' ',text).strip()
  return text
dataset=Dataset.from_dict({'text':[raw_text]})
cleaned_dataset=dataset.map(lambda x: {'text':clean_text(x['text'])})
print('Cleaned Text:',cleaned_dataset[0]['text'])
```

```python
from transformers import pipeline
sentiment_analyzer=pipeline('sentiment-analysis')
text="I love working with Hugging Face! It's amazing."
result=sentiment_analyzer(text)
print('Sentiment Analysis Result:',result)
```
